```python
basiclogs_las = []
interpretedlogs_las = []
for las_file in las_files:
    temp_las_file = lasio.read(las_file)
    temp_df = temp_las_file.df()
    well_name = temp_las_file.well[5]['value']
    temp_df['Well_Name'] = well_name
    if temp_df.shape[1] == 6:
        basiclogs_las.append(temp_df.reset_index())
    else:
        interpretedlogs_las.append(temp_df.reset_index())
```

```python
print(basiclogs_las)
```
